```python
import pandas as pd
from scipy.signal import find_peaks
import plotly.graph_objects as go
```

```python
xrd_data = pd.read_csv('S2_1024.xy', delimiter=' ', names=['Angle', 'Intensity'])

indices = find_peaks(xrd_data.Intensity)[0]

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=xrd_data.Angle,
    y=xrd_data.Intensity,
    mode='lines+markers',
    name='XRD sample'
))

fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=xrd_data.Angle,
    y=xrd_data.Intensity,
    mode='lines+markers',
    name='XRD sample'
))

fig.add_trace(go.Scatter(
    x=xrd_data.Angle.loc[indices],
    y=xrd_data.Intensity.loc[indices],
    mode='markers',
    marker=dict(
        size=8,
        color='red',
        symbol='cross'
    ),
    name='Detected Peaks'
))

fig.show()
```
